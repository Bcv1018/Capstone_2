## Core Marketing Analysis

### Who are the territory managers for the sales territories assigned? What are the store IDs and cities for the stores in each assigned sales territory?

```python
state_info = store_detail[(store_detail['State'] == 'Texas') | (store_detail['State'] == 'South Carolina')] # Filters store_detail df to only show rows that have either Texas or South Carolina
territory_info = state_info[['Store ID','Store Location','State','Territory Manager']] # creates a new df that will only show the columns that are relevant right now
territory_info.set_index('Store ID', inplace=True) # sets Store ID to be the index of the territory_info df

# Calls territory_info DF 
territory_info
# Territory Manager for Texas is Jeff "Howdy" Richards and for South Carolina is Len Jensen
# Store I
```

### What is monthly total revenue for in-store sales in each of the two sales territories, over the full period covered by the data?

```python
# Merging store sales with territory_info to create a new df 
territory_sales = pd.merge(territory_info, store_sales, on=['Store ID'])

territory_sales
```

```python
# Converting transaction date from str to an actual date value and adding new date column that is only monthly
territory_sales['Transaction Date'] = pd.to_datetime(territory_sales['Transaction Date'])
territory_sales['Date'] = territory_sales['Transaction Date'].dt.to_period('M')

territory_sales
```

```python
# monthy revenue for Texas and South Carolina over the full period covered by the data which is from 2022-01 to 2025-12
monthly_revenue = territory_sales.groupby(['State','Date'])[['Sale Amount']].sum()

monthly_revenue
```

### How would you rank the sales performance of each store in each sales territory? Which are the top-performing stores?

```python
# Creating a new df that is filtered down to just state storeID storeLocation by Total Revenue
total_rev = territory_sales.groupby(['State','Store ID','Store Location'])[['Sale Amount']].sum().round(0).astype(int)
total_rev.rename(columns={'Sale Amount':'Total Revenue'},inplace=True)


total_rev
```

```python
# Sorting total_rev in descending order to get the most revenue at the top and least revenue at the bottom
total_rev = total_rev.groupby('State',group_keys=False).apply(lambda x: x.sort_values('Total Revenue', ascending=False))
total_rev
```

```python
# creating a ranking system for each Territory
total_rev['Rank'] = total_rev.groupby('State',group_keys=False).apply(lambda x:x['Total Revenue'].rank(ascending=False).astype(int))
total_rev
```

The top performing store for South Carolina is Greenville with $337,002
The top performing store for Texas is Beaumont with $637,819

###  Comparing the customer ID from the customer list data with the rewards ID from the sales data, who were the top customers in each sales territory?

```python
# Merging territory_sales df with customer list to make a new df that
rewards_info = pd.merge(territory_sales,cx_list, left_on='RewardsID', right_on='cust_id')
rewards_info
```

```python
# Filtered rewards_info into a new df only having the columns that we need
rewards_mem = rewards_info[['State','cust_id','name','Sale Amount']]
rewards_mem
```

```python
# Finding the total sale amount for each customer
total_cx_spent = rewards_mem.groupby(['State','cust_id','name'])[['Sale Amount']].sum().round(0).astype(int) # Filters down to just state,custid,name,and sale amount while also finidng the total sale amout per customer
total_cx_spent.rename(columns={'Sale Amount' : 'Total Spent'},inplace=True) # Just renames Sale Amount to Total Sales


total_cx_spent
```

```python
# Ordering the total spent by descending order and only showing top ten for each state
total_cust_spent = total_cx_spent.groupby('State',group_keys=False).apply(lambda x: x.sort_values('Total Spent', ascending=False)) # Orders customers from descending order based on how much they spent seperated by state 
top_10_cust = total_cust_spent.groupby('State', group_keys=False).apply(lambda x: x.head(10)) # only shows the top ten customers based on their total spent seperated by state

top_10_cust
```

### What is the number of transactions per month by product category in each assigned territory?

```python
# Merging Products and ProductCategories
prod_details = pd.merge(products,prod_cats, on=['CategoryID','SubcategoryID'])
prod_details
```

```python
# Merging prod_details and StoreSales
prod_sales = pd.merge(prod_details,territory_sales, on='Prod Num')
prod_sales
```

```python
# cleaning up prod sales to get ready for monthly transaction per category
cat_monthly = prod_sales[['State','Date','Category','Sale Amount']]
cat_monthly
```

```python
# Counting the amount of transaction monthly
cat_monthly_transaction = cat_monthly.groupby(['State','Date','Category'])[['Sale Amount']].count()
cat_monthly_transaction
```

### What is total sales revenue per month by category?

```python
# Total sales revenue per month by category
cat_monthly_revenue = cat_monthly.groupby(['State','Date','Category'])[['Sale Amount']].sum().round(0).astype(int)
cat_monthly_revenue
```

```python
# Seperating texas and south carolina to two different df
texas_cat = cat_monthly_revenue.xs('Texas', level='State')
sc_cat = cat_monthly_revenue.xs('South Carolina', level='State')

# Reseting the indexes to make them into columns
texas_cat = texas_cat.reset_index().pivot_table(index='Date', columns='Category',values='Sale Amount')
sc_cat = sc_cat.reset_index().pivot_table(index='Date',columns='Category',values='Sale Amount')

# Filtering the charts to only include 2025
texas_2025 = texas_cat[texas_cat.index.year == 2025]
sc_2025 = sc_cat[sc_cat.index.year == 2025]
```

```python
texas_2025
```

```python
sc_2025
```

```python
# Creating subplot for cat_monthly_revenue
fig, (ax1,ax2) = plt.subplots(1,2, figsize=(14,4))

texas_2025.plot(kind='bar', ax=ax1)
ax1.set_title('Texas Monthly Revenue by Category')
ax1.set_ylim(0, 22000)
ax1.get_legend().remove()
ax1.set_ylabel('Revenue ($)')

sc_2025.plot(kind='bar', ax=ax2)
ax2.set_title('South Carolina Monthly Revenue by Category')
ax2.set_ylim(0,6000)
ax2.get_legend().remove()
ax2.set_ylabel('Revenue ($)')


fig.text(.5,-.02,'Note: Technology exceeds chart scale',ha='center', fontsize=15)

handles, labels = ax1.get_legend_handles_labels()
fig.legend(handles, labels, loc='lower center',ncol=6, bbox_to_anchor=(.5,-.5))

# plt.tight_layout()
plt.subplots_adjust(bottom=.25, top=.95)
plt.savefig('2025_Monthly_Revenue.jpg',bbox_inches='tight', pad_inches=.5)
plt.show()
```
